**Sales Analysis Using Python (Pandas & Matplotlib)**

*Project Title*

***Product-wise Monthly Sales Analysis***

 Problem Statement
A retail shop sells different products every month. The sales manager wants to understand:

•	Which product sells the most

•	Monthly sales trends

•	Product-wise sales comparison



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_excel(r"C:/Users/RAJESH MALIK/OneDrive/Desktop/python folder/sales_data.xlsx")
print(data.head(15))
```

```python
product_sales= data.groupby('Product')['Sales'].sum()
print(product_sales)
```

```python
product_sales.plot(kind='bar',color='purple')
plt.title('Product-wise total sales')
plt.xlabel('Product')
plt.ylabel('Sales Amount')
plt.show()
```

```python
monthly_sales= data.groupby('Month')['Sales'].sum()
monthly_sales.plot(kind='line',color='pink')
plt.title('Month wise Sales')
plt.xlabel('Month')
plt.ylabel('Sale Amount')
plt.show()
```

```python
product_sales.plot(kind='pie',autopct='%1.1f%%')
plt.title('Sales contribution by product')
plt.show()
```

**Observations**

•	*Laptop has the highest sales among all products*

•	*Sales increase from January to March*

•	*Headphones contribute the least to total sales*
